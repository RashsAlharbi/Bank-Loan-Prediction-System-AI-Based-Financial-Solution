```python
#import
import numpy as np #For numerical calculation, array
import pandas as pd #For data manipulation
import matplotlib.pyplot as plt #is used for plotting graphs.
%matplotlib inline
import sklearn #is a free machine learning library for Python
from sklearn.model_selection import train_test_split# to test and train dataset
from sklearn.metrics import confusion_matrix # to define the TP,FP,FN,TN
#machine learning algorithms and functions 
from sklearn.tree import DecisionTreeClassifier
import sklearn.metrics as metrics
from sklearn.metrics import accuracy_score
# IMPORT FOR PRINT TABLE REQUIRMENT (recall, precision, accuracy, matrix)
from sklearn.metrics import classification_report
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn import tree
```

```python
#Reading the dataset file
from google.colab import drive
drive.mount('/content/gdrive')
dataset = pd.read_csv('gdrive/My Drive/AI Labs/Bank.csv')
dataset.head()
```

```python
dataset.shape
```

```python
X = dataset.values[:, 0:12]
Y = dataset.CreditCard
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=100)
```

```python
# Decision tree classifier
clf= DecisionTreeClassifier(random_state = 100)
#fitting the training data
clf.fit(X_train, Y_train)
```

```python
Y_pred = clf.predict(X_test)# use the features only and predect the outcom 
print('\n --------------------------The PREDICTION values-------------------------- \n')
print(Y_pred)
```

```python
print('\n ----------The Accuracy is---------- \n')
print(metrics.accuracy_score(Y_test, Y_pred))
print(' so ... ',metrics.accuracy_score(Y_test, Y_pred)*100,'%')
#Matrix
print('\n ----------The Matrix is---------- \n')
cm=np.array(confusion_matrix(Y_test,Y_pred))
cm
```

```python
print(classification_report(Y_test,Y_pred))
```

```python
tree.plot_tree(clf)
```
